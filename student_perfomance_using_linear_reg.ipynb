```python
##checkinf for outliers
def detect_outliers(df, columns): # Pass the DataFrame to the function
  outliers_dict={}
  for col in columns:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    # Iterate through the values of the current column, not the entire DataFrame
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].tolist()
    outliers_dict[col] = outliers # Store outliers for each column in the dictionary
  return outliers_dict
numeric_columns = df.select_dtypes(include=['number']).columns # Select numeric columns
outliers = detect_outliers(df, numeric_columns) # Pass numeric columns to the function
print("Outliers:", outliers)
```

```python
x_cols=df3[['Hours Studied', 'Extracurricular Activities',
       'Sleep Hours', 'Sample Question Papers Practiced']]
x_cols.head()
```

```python
y_cols=df3['Performance Index']
y_cols.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_cols,y_cols,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lregr=LinearRegression()
lregr.fit(X_train,y_train)
```

```python
y_pred=lregr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse=np.sqrt(mean_squared_error(y_test,y_pred))
mse
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
error=np.sqrt(mean_squared_error(y_test,y_pred))
error
```
